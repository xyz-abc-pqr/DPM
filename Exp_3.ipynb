Calculating the Total Null Values in the dataset

```python
df.isna().sum()
```

Checking where are the Null Values

```python
df[df.isnull().any(axis=1)]
```

### Creating the Heatmaps to see the null values

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(df.isnull())
plt.title("Null Values in the Columns")
plt.show()
```

### Dropping the Null Values where all the rows have the null values

```python
df.dropna(axis=0,how="all",inplace=True)
```

Again Checking the Null Values

```python
df.isna().sum()
```

Checking Where are the Null Values

```python
df[df.isnull().any(axis=1)]
```

```python
df.reset_index(inplace=True)
```

## Fill Null Values with 0

```python
df1=df.fillna(0)
```

```python
df1
```

```python
df1.isna().sum()
```

## Fill Null Values with Forward Fill method

```python
df.fillna(method="ffill")
```

## Fill Null Values with backward fill method

```python
df.fillna(method="bfill")
```

## Fill Null Values with Interpolate

```python
df2=df.interpolate()
```

```python
df2
```

```python
df2.isna().sum()
```

```python
df2.drop_duplicates(keep="first")
```
