## Extracting the Top Level Comments from the above Video

Calling the Youtube API to extract the comments

```python
comment_url=url+"/commentThreads?key={}&part=snippet&videoId={}&maxResults=30".format(api_key,video_id)
comment_url
```

```python
commentData=requests.get(comment_url)
commentData
```

```python
commentData=commentData.json()
commentData
```

Extracting the Comments

```python
comments=[item["snippet"]["topLevelComment"]["snippet"]["textOriginal"] for item in commentData["items"]]
comments
```

Displaying the Comments

```python
df=pd.DataFrame({"Comments":comments})
df
```
